```python
import numpy as np
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.utils.validation import check_X_y, check_array, check_is_fitted
from sklearn.utils.multiclass import unique_labels
from sklearn.metrics import euclidean_distances
from tensorflow.keras import models, layers, utils, backend as K
from tensorflow.random import set_seed
from tensorflow.keras.optimizers import SGD

class TemplateClassifier(BaseEstimator, ClassifierMixin):
    
    def __init__(self, dropoutrate = 0.5, learning_rate=0.0001, 
                 momentum=0.8, batchsize=16, numepochs = 100,
                decay_rate = 0.00001):
        
        self.dropoutrate = dropoutrate
        self.learning_rate = learning_rate
        self.momentum = momentum
        self.batchsize = batchsize
        self.numepochs = numepochs
        self.decay_rate = decay_rate
    
    
    def fit(self, X, y):

# Define model architecture
        model = models.Sequential(name="SimpleNN", layers=[
            layers.Dense(name="h1", input_dim=3,
                 units=5, activation='relu'),
            layers.Dropout(name="drop1", rate=self.dropoutrate, seed=0),

            layers.Dense(name="h2", units=5, activation='relu'),
            layers.Dropout(name="drop2", rate=self.dropoutrate, seed=1),
            layers.Dense(name="output", units=1, activation='linear')
        ])
    
        def R2(y, y_hat):
            ss_res =  K.sum(K.square(y - y_hat)) 
            ss_tot = K.sum(K.square(y - K.mean(y))) 
            return ( 1 - ss_res/(ss_tot + K.epsilon()) )
        
        sgd = SGD(lr=self.learning_rate, momentum=self.momentum, decay=self.decay_rate, nesterov=True)
        
        model.compile(optimizer=sgd, loss='mean_squared_error', metrics=[R2])
        
        model.fit(x=X, y=y, 
                  batch_size=self.batchsize, epochs=self.numepochs, shuffle=True, verbose=0,
                  validation_split=0.3)
        
        self.__dict__.update(model.__dict__)

        return self

    def predict(self, X):

# Check if fit has been called
        #check_is_fitted(self)

# Input validation
  #      X = check_array(X)
        return self.predict(X)

   #     closest = np.argmin(euclidean_distances(X, self.X_), axis=1)
    #    return self.y_[closest]
```
